```python
# FOR NUMERICAL ANALYTICS
import numpy as np
# TO STORE AND PROCESS DATA IN DATAFRAME
import pandas as pd
import os
# BASIC VISUALIZATION PACKAGE
import matplotlib.pyplot as plt
# ADVANCED PLOTING
import seaborn as sns
```

```python
#2019 data
df = pd.read_csv('2019.csv')
df = df[['Score','GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices','Generosity','Perceptions of corruption']]
sns.displot(df, x="Score",binwidth=0.4,kde=True)
plt.title('2019', fontsize=16)
plt.savefig('2019_Score.jpg', bbox_inches='tight')
corr = df.corr()
mask = np.zeros_like(corr)
cmap = sns.diverging_palette(230, 10, as_cmap=True)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(7, 7))
    ax = sns.heatmap(corr, mask=mask,cmap=cmap, vmin=-1, vmax=1, square=True, annot=True)
plt.title('2019', fontsize=16)
plt.savefig('2019_Corr.jpg', bbox_inches='tight')
```

```python
df2015 = pd.read_csv('2015.csv')[['Country','Happiness Rank']]
df2015.columns = ['Country name','2015 Rank']
df2016 = pd.read_csv('2016.csv')[['Country','Happiness Rank']]
df2016.columns = ['Country name','2016 Rank']
df2017 = pd.read_csv('2017.csv')[['Country','Happiness.Rank']]
df2017.columns = ['Country name','2017 Rank']
df2018 = pd.read_csv('2018.csv')[['Country or region','Overall rank']]
df2018.columns = ['Country name','2018 Rank']
df2019 = pd.read_csv('2019.csv')[['Country or region','Overall rank']]
df2019.columns = ['Country name','2019 Rank']
```

```python
df2020 = pd.read_csv('2020.csv')
df2020['Rank'] = df2020["Ladder score"].rank(ascending=False).astype(int)
df2020 = df2020[['Country name','Rank']]
df2020.columns = ['Country name','2020 Rank']
```

```python
df2021 = pd.read_csv('2021.csv')
df2021['Rank'] = df2021["Ladder score"].rank(ascending=False).astype(int)
df2021 = df2021[['Country name','Rank']]
df2021.columns = ['Country name','2021 Rank']
```

```python
res = pd.merge(left=df2015, right=df2016, left_on='Country name', right_on='Country name')
res = pd.merge(left=res, right=df2017, left_on='Country name', right_on='Country name')
res = pd.merge(left=res, right=df2018, left_on='Country name', right_on='Country name')
res = pd.merge(left=res, right=df2019, left_on='Country name', right_on='Country name')
res = pd.merge(left=res, right=df2020, left_on='Country name', right_on='Country name')
res = pd.merge(left=res, right=df2021, left_on='Country name', right_on='Country name')
res.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
#from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
import scipy.stats as ss
import itertools
sns.set()

%matplotlib inline

labels = res['Country name'].index
#labels = pd.get_dummies(labels)
features = res.loc[:, res.columns != 'Country name']
features = pd.get_dummies(features)
features_train, features_test, labels_train, labels_test = \
        train_test_split(features, labels, test_size=0.2)
model = DecisionTreeRegressor()
    
    # Train it on the **training set**
model.fit(features_train, labels_train)
    # Compute training accuracy
train_predictions = model.predict(features_train)
    # Compute test accuracy
test_predictions = model.predict(features_test)
print('Train Error:', mean_squared_error(labels_train, train_predictions))
print('Test Error:', mean_squared_error(labels_test, test_predictions))
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
X=res.loc[:, res.columns != 'Country name']
y=res['Country name'].index
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=1)
regr = MLPRegressor(random_state=1, max_iter=500).fit(X_train, y_train)
prediction = regr.predict(X_test[:2])
print(regr.score(X_test, y_test))
np_array = regr.predict(X)
new_series = pd.Series(np_array)
res['2022 predict rank'] = new_series
res.to_csv('prediction.csv',index=False)
res.head(10)
```

```python

# Define a function to plot the accuracies
accuracies = []
for i in range(1, 30):
    X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=i)
    regr = MLPRegressor(random_state=i, max_iter=500).fit(X_train, y_train)
    prediction1 = regr.predict(X_train[:2])
    score_train = regr.score(X_train, y_train)
    prediction2 = regr.predict(X_test[:2])
    score_test = regr.score(X_test, y_test)


    accuracies.append({'random_state': i, 'score_train': score_train, 
                       'score_test': score_test})
accuracies = pd.DataFrame(accuracies)


column = 'score_train'
name = 'Train'
sns.relplot(kind='line', x='random_state', y=column, data=accuracies)
plt.title(f'{name} Score as Max Depth Changes')
plt.xlabel('random_state')
plt.ylabel(f'{name} Score')
plt.ylim(0, 1)

plt.show()  # Display the graph

column = 'score_test'
name = 'Test'
sns.relplot(kind='line', x='random_state', y=column, data=accuracies)
plt.title(f'{name} Score as Max Depth Changes')
plt.xlabel('random_state')
plt.ylabel(f'{name} Score')
plt.ylim(0, 1)

plt.show()  # Display the graph
```
